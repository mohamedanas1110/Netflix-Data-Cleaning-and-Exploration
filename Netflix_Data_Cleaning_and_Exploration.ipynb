```python
df.duplicated().sum()
```

```python
df['director'].replace(np.nan, 'No Data', inplace=True)


df['cast'].replace(np.nan, 'No Data', inplace=True)
```

```python
categorical_column = ['type']
df['categorical_column'] = df[categorical_column].astype('category')
```

```python
df.head()
```

```python
#5
df.isnull().sum()
```

```python
df['director'].replace(np.nan, 'No Data', inplace=True)
df['cast'].replace(np.nan, 'No Data', inplace=True)
```

```python
#5)
df.isna()
```

```python
#5)
df.isnull().sum()
```

```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
```

```python
df.head()
```

```python
df.isna()
```

```python
df.isnull().sum()
```

```python
movie_rating = df.loc[df['type'] == 'Movie', 'rating'].mode()[0]
tv_rating = df.loc[df['type'] == 'TV Show', 'rating'].mode()[0]

df['rating'] = df.apply(lambda x: movie_rating if x['type'] == 'Movie' and pd.isna(x['rating'])
                        else tv_rating if x['type'] == 'TV Show' and pd.isna(x['rating'])
                        else x['rating'], axis=1)
```

```python
df.isna()
```

```python
df.isnull().sum()
```

```python
movie_duration_mode = df.loc[df['type'] == 'Movie', 'duration'].mode()[0]
tv_duration_mode = df.loc[df['type'] == 'TV Show', 'duration'].mode()[0]

df['duration'] = df.apply(lambda x: movie_duration_mode if x['type'] == 'Movie'
                          and pd.isna(x['duration'])
                          else tv_duration_mode if x['type'] == 'TV Show'
                          and pd.isna(x['duration'])
                          else x['duration'], axis=1)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
#4) Visual Analysis - Univariate,
#Bivariate after pre-processing of the data


plt.figure(figsize=(8,6))
sns.countplot(x='type',data=df)
plt.title('Distribution of Movies & TVShows')
plt.show()
```

```python
df["date_added"] = pd.to_datetime(df['date_added'])
```

```python
df.head()
```

```python
df.head()
```

```python
df['cast'] = df['cast'].str.split(', ')
df['director'] = df['director'].str.split(', ')
df['country'] = df['country'].str.split(', ')
```

```python
df = df.explode('cast').explode('director').explode('country')
```

```python
plt.figure(figsize=(8,6))
sns.distplot(df['release_year'],kde=False, bins=30, color='blue')
plt.title('Distribution of movies across release years')
plt.xlabel('release_year')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='duration', data=df, palette='viridis')
plt.title('Distribution of movies by duration')
plt.xlabel('duration')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['duration'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of movies by duration')
plt.xlabel('duration')
plt.ylabel('count')
plt.show()
```
